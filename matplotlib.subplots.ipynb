```python
fig,axes=plt.subplots(nrows=2,ncols=2)
axes[0][0].plot(x,y)
axes[0][1].plot(x,y)
axes[1][0].plot(x,y)
axes[1][1].plot(a,b)
fig.subplots_adjust(wspace=0.9)
```

```python
fig,axes=plt.subplots(nrows=2,ncols=2)
axes[0][0].plot(x,y)
axes[0][1].plot(x,y)
axes[1][0].plot(x,y)
axes[1][1].plot(a,b)
fig.subplots_adjust(wspace=0.9,hspace=1)
```

```python
fig,axes=plt.subplots(nrows=2,ncols=2,figsize=(4,10),dpi=200)

axes[0][0].plot(x,y)

axes[0][1].plot(x,y)

axes[1][0].plot(x,y)
axes[1][0].set_ylabel('Y label')

axes[1][1].plot(a,b)
axes[1][1].set_title('TITLE 1.1')

plt.tight_layout()

fig.savefig('new_subplots.png',bbox_inches='tight')
```
